```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
%matplotlib inline
plt.style.use('ggplot')
mpl.rcParams['axes.unicode_minus']=False
```

```python
import os
os.chdir('/Users/zoe/Downloads/data')
```

```python
train=pd.read_csv("train.csv", parse_dates=["datetime"])
train.shape
```

```python
train.info()
```

```python
train.head()
```

```python
train["year"]=train["datetime"].dt.year
train["month"]=train["datetime"].dt.month
train["day"]=train["datetime"].dt.day
train["hour"]=train["datetime"].dt.hour
train.head()
```

```python
fig1, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2)
fig1.set_size_inches(10,10)

sns.barplot(data=train, x="year", y="count", ax=ax1)
sns.barplot(data=train, x="month", y="count", ax=ax2)
sns.barplot(data=train, x="day", y="count", ax=ax3)
sns.barplot(data=train, x="hour", y="count", ax=ax4)

ax1.set(ylabel='Count', xlabel='year', title="yearly count")
ax2.set(xlabel='month', title="monthly count")
ax3.set(xlabel='day', title="daily count")
ax4.set(ylabel='hour', title="hourly count")
```

```python
fig2, axes=plt.subplots(nrows=2,ncols=2)
fig2.set_size_inches(10,10)
sns.boxplot(data=train, y="count",orient="v",ax=axes[0][0])
sns.boxplot(data=train, y="count", x="season", orient="v", ax=axes[0][1])
sns.boxplot(data=train, y="count", x="hour", orient="v", ax=axes[1][0])
sns.boxplot(data=train, y="count", x="workingday", orient="v", ax=axes[1][1])

axes[0][0].set(ylabel='Count', title='Count')
axes[0][1].set(xlabel='Season', title="seasonal count")
axes[1][0].set(xlabel='Hour of the day', title="hourly count")
axes[1][1].set(xlabel='working day', title="count for working day")
```
