# Description
- ID - an Id that represents a (Shop, Item) tuple within the test set
- shop_id - unique identifier of a shop
- item_id - unique identifier of a product
- item_category_id - unique identifier of item category
- item_cnt_day - number of products sold. You are predicting a monthly amount of this measure
- item_price - current price of an item
- date - date in format dd/mm/yyyy
- date_block_num - a consecutive month number, used for convenience. January 2013 is 0, February 2013 is 1,..., October 2015 is 33

```python
items = pd.read_csv('fs/items.csv')
items.head()
```

```python
# item category 합치기
newtrain = pd.merge(items,train,on='item_id')
newtrain.head()
```

```python
# 동일 shop_id, item_id 별로 item_cnt_day 합하기 
grouped = pd.DataFrame(newtrain.groupby(['shop_id','item_category_id','date_block_num'])
                       ['item_cnt_day'].sum().reset_index())
grouped.rename(columns={'item_cnt_day':'item_cnt_mon'},inplace=True)
grouped.head(30)
grouped1= grouped = pd.DataFrame(newtrain.groupby(['shop_id','item_id','date_block_num'])
                       ['item_cnt_day'].sum().reset_index())
grouped1.rename(columns={'item_cnt_day':'item_cnt_mon'},inplace=True)
```

```python
mean = pd.DataFrame(grouped1.groupby(['shop_id','item_id'])
                    ['item_cnt_mon'].sum().reset_index())
mean.rename(columns={'item_cnt_mon':'total_item_cnt'},inplace=True)
mean['mean_item_cnt'] = mean['total_item_cnt']/33
mean.head()
```
